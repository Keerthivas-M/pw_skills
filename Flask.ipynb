Q3. What is App routing in Flask? Why do we use app routes?

App Routing means mapping the URLs to a specific function that will handle the logic for the URL. Morder web frameworks use more meaningful URLs to help users remeber the URLs and make navigation simpler. 
The URL ("/") is associated with the roor URL. For Example, 'http:/www.someurl.com' is a URL and adding routing "/hello" in application would be looked like "http:/www.someurl.com/hello".  

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999Capture.PNG

Capture.PNG

```python
from flask import Flask

app =  Flask(__name__)

@app.route("/")
def func1():
    return "Company Name: ABC Corporation\nLocation:India\nContact Detail:999-999-9999"

@app.route("/welcome")
def welcome_func():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host='0.0.0.0')
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
from flask import Flask

app =  Flask(__name__)

@app.route("/")
def func1():
    return "Company Name: ABC Corporation\nLocation:India\nContact Detail:999-999-9999"

@app.route("/welcome")
def welcome_func():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host='0.0.0.0')
```
