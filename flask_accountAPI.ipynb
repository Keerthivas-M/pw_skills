Q1. Explain GET and POST methods.

GET method is used to request data from a specified resource. It is not a secure, it can be cached, remain in the browser history. IT adds the data to the URL.

POST method is used to send data to a server to create/update a resource. It can not be cached, not visible in URL.

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/login', methods=['POST'])
def user_login():
  if (request.method == 'POST'):
    username = request.json['username']
    pwrd = request.json['password']
    db = {'peter':'peter123'}
    if username in db:
      if pwrd == db[username]:
        result = "You're logged into your account"
      else:
        result = "Incorrect username and password combination"
    else:
      result = "You dont have a account with this username"
  
  return jsonify(result)

if __name__=="__main__":
  app.run(host="0.0.0.0")
```
